```python
import matplotlib.dates as mdates
fig, ax = plt.subplots(figsize=(10,5))

ax.set_title("Pierwsza wykres")
ax.set_xlabel("Czas", rotation=45)
ax.set_ylabel("Sprzedaz")

plt.xticks(rotation=45, ha='right')

xformatter = mdates.DateFormatter('%H:%M')
plt.gcf().axes[0].xaxis.set_major_formatter(xformatter)


## to jest z series sprzedaz
ax.plot(dane['DT'], dane['Sprzedaz'])
for i,j in zip(dane['DT'],dane['Sprzedaz']):
    ax.annotate(str(j), xy=(i,j))

## to jest z series sprzedaz2
ax.plot(dane['DT'], dane['Sprzedaz2'], linewidth= 8.0, linestyle='--', color = "g")
for i,j in zip(dane['DT'],dane['Sprzedaz2']):
    ax.annotate(j, xy=(i,j))

ax.legend(["Sprzedaz", "Sprzedaz2"])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

for bars in ax.containers:
  ax.bar_label(bars)

plt.show()
```

```python
df = pd.DataFrame({
    'Male': [35, 20, 43, 12, 19],
    'Female': [20, 11, 18, 6, 9],
    "Don't know" : [20, 11, 18, 6, 9]
 })

Class = ["Fisrt","Second","Third","Fourth","Fifth"]
ax = df.plot(stacked=True, kind="box")
```

```python
ax = df.plot(stacked=True, kind="bar")
for bar in ax.patches:
  print(bar)
  height = bar.get_height()
  width = bar.get_width()
  
  x = bar.get_x()
  y = bar.get_y()

  label_text = height
  label_x = x + width/2
  label_y = y + height/2

  ax.text(label_x, label_y, label_text, ha="center", va="center", color="white")

  ax.set_xticklabels(Class, rotation="horizontal")
  
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.3, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig, ax = plt.subplots()
ax.pie(sizes, labels = labels, explode= explode, radius=0.9,  autopct= "%1.0f%%")

plt.plot()
```

```python
fig, ax = plt.subplots()

labels = ['label1', 'label2', 'label3']
size = 0.3
vals = np.array([[60., 32.], [37., 40.], [29., 10.]])

cmap = plt.colormaps["tab20c"]
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap([1,2,5,6,9,10])

ax.pie(vals.sum(axis=1), labels = labels, radius=1, colors= outer_colors, wedgeprops = dict(width=size, edgecolor="w")) #dokonujemy sumy 3 zestaww

ax.pie(vals.flatten(), radius=1-size, colors= inner_colors, wedgeprops = dict(width=size, edgecolor="w")) #naoylimy 

ax.set(aspect="equal", title="Second pie")
```
